```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exercise 1: a) and b)
# Plotting the signals
fig, axs = plt.subplots(3)
fig.suptitle("Semnalele x(t), y(t), z(t)")

x = lambda t : np.cos(520*np.pi*t + np.pi/3)
y = lambda t : np.cos(280 *np.pi*t - np.pi/3)
z = lambda t : np.cos(120 *np.pi*t + np.pi/3)

time = np.linspace(0, 0.03, 200)

axs[0].plot(time, x(time))
axs[1].plot(time, y(time))
axs[2].plot(time, z(time))
plt.show()
```

```python
# Exercise 1: c)
fig, axs = plt.subplots(3)
fig.suptitle("Semnalele x[t], y[t], z[t] esantionate cu o frecventa de 200Hz")

samples = np.linspace(0, 1, num=200)
axs[0].stem(samples, x(samples))
axs[0].plot(time, x(time))
axs[1].stem(samples, y(samples))
axs[1].plot(time, y(time))
axs[2].stem(samples, z(samples))
axs[2].plot(time, z(time))

axs[0].set_xlim([0,0.03])
axs[1].set_xlim([0,0.03])
axs[2].set_xlim([0,0.03])
plt.show()
```

```python
# Exercise 2: a)
fig, axs = plt.subplots(2)
fig.suptitle("Un semnal sinusoidal de frecventa 400Hz, care sa contina 1600 de esantioane")
fig.tight_layout()

x = lambda t: np.sin(2*np.pi*t*400)
time = np.linspace(0, 1, num= 1600)

axs[0].set_title("Pe intervalul [0,1]")
axs[0].stem(time, x(time))


lod = np.linspace(0, 0.03, num=10000)
axs[1].stem(time, x(time))
axs[1].plot(lod, x(lod))
axs[1].set_xlim([0,0.03])
axs[1].set_title("Pe intervalul [0,0.03]")
plt.show()

```

```python
# Exercise 2: b)

fig, axs = plt.subplots(3)
fig.suptitle("Un semnal sinusoidal de frecventa 800Hz care sa dureze 3 secunde")
x = lambda t: np.sin(2*np.pi*800*t)
fig.tight_layout()

lod_1 = np.linspace(0,3, 100)
lod_2 = np.linspace(0,3, 1000)
lod_3 = np.linspace(0,3, 10000)

axs[0].plot(lod_1, x(lod_1))
axs[0].set_title("Samples = 10^2")
axs[1].plot(lod_2, x(lod_2))
axs[1].set_title("Samples = 10^3")
axs[2].plot(lod_3, x(lod_3))
axs[2].set_title("Samples = 10^4")
plt.show()
```

```python
# Exercise 2: c)

fig, axs = plt.subplots(1)
fig.suptitle("Un semnal de tip sawtooth de frecventa 240Hz")

frequency = 240 
period = 1/frequency

time = np.linspace(0,0.05, 100000)
x = lambda t: 2*(t/period- np.floor(1/2+t/period)) 

axs.plot(time, x(time))
plt.show()
```

```python
# Exercise 2: d)

fig, axs = plt.subplots(1)
fig.suptitle("Un semnal de tip square de frecventa 300Hz")

frequency = 300 
x = lambda t: 2*(2*np.floor(frequency*t) - np.floor(2* frequency*t)) + 1 
time = np.linspace(0, 0.03, 1000)
axs.plot(time, x(time))

plt.show()
```

```python
# Exercise 2: f)

fig, axs = plt.subplots(1)
fig.suptitle("Un semnal 2D de dimensiune 128x128")



def mandelbrot(c,maxiter):
    z = c
    for n in range(maxiter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return 0

width = 128
height = 128

coord_x = np.linspace(-2, 1, width)
coord_y = np.linspace(-1.5, 1.5, height)

signal = np.zeros((width,height))
for i in range(height):
    for j in range(width):
        signal[i,j] = mandelbrot(coord_x[j]+1j*coord_y[i], 20)

plt.imshow(signal)
plt.show()
```
